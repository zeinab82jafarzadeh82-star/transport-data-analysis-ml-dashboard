```python
#LINEAR REGRESSION on transport_df dataset

#importing needed libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#برای اینکه روی دیتاست الگوریتم رگرسیون خطی را پیاده سازی کنیم باید
#ستون های عددی را به کار بگیریم ،
#اجازه استفاده از ستون های غیرعددی را نداریم
#باید ستون هدف را از بین سایر ستون های عددی حذف کنیم چون قرار است آن را پیش بینی کنیم
#transport_df_numercal=transport_df.select_dtypes(include=['int','float'])
transport_df_numercals = transport_df.select_dtypes(include=['int', 'float']).drop('Calculateion fare', axis=1)

#transport_df_numercals= transport_df_numercal.drop('Calculateion fare', axis=1).values
X = transport_df_numercals
#ستون ویژگی کرایه محاسباتی ، هدف است
y=transport_df['Calculateion fare']


#تقسیم داده به آموزش و تست
# 30% برای تست
#70% برای آموزش
X_train , X_test , y_train ,y_test= train_test_split(X , y ,test_size=0.3,random_state=42)

#ساخت مدل رگرسیون خطی
model=LinearRegression()
#فیت کردن وآموزش مدل
model.fit(X_train,y_train)

#ذخیره ضرایب مدل برای رسم نمودار اهمیت
importance = model.coef_
#ذخیره ستون های
#X_train
#در دیتافریم
feature_names = X_train.columns

#پیش بینی و ارزیابی مدل
y_pred=model.predict(X_test)

print("Mean Absolute Error (MAE):", metrics.mean_absolute_error(y_test,y_pred))
print("Mean Squared Error (MSE):",metrics.mean_squared_error(y_test,y_pred))
print("R2 Score :",metrics.r2_score(y_test,y_pred))

#رسم نمودار برای درک بهتر
#scatterplot()
#این نمودار را برای مقایسه مقادیر واقعی و مقادیر پیش بینی شده رسم می کنیم
plt.figure(figsize=(10,4))
plt.title("Actual vs Predicted Calculation Fare")
plt.xlabel('Actual Fare (y_test)')
plt.ylabel("Predicted Fare (y_pred)")
sns.scatterplot(x=y_test,y=y_pred)
plt.grid(True)
plt.show()


#نمودار خطاها
#اختلاف بین مقادیر واقی و پیش بینی شده را بررسی می کنیم
residuals= y_test - y_pred

plt.figure(figsize=(10,4))
sns.scatterplot(x=y_pred,y=residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.title("residuals vs y_pred")
plt.xlabel("Predicted Fare")
plt.ylabel("Residuals (Errors)")
plt.grid(True)
plt.show()

#نمودار هیستوگرام خطاها
#می خواهیم ببینیم توزیع خطاها به چه صورتی است
plt.figure(figsize=(10,4))
sns.histplot(x=residuals,bins=30)
plt.title("Distribution of Prediction Errors (Residuals)")
plt.xlabel("Error")
plt.ylabel("Count")
plt.grid(True)
plt.show()

# ساخت یک دیتافریم مرتب‌شده بر اساس اهمیت
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importance
}).sort_values(by='Importance', key=abs, ascending=False)

#نمودار اهمیت ویزگی ها
plt.figure(figsize=(10,6))
sns.barplot(x='Importance', y='Feature', data=importance_df)
plt.title('Feature Importance (Linear Regression Coefficients)')
plt.xlabel('Coefficient Value')
plt.ylabel('Feature')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#SVM algorithm on transport_df dataset

#چون برای این الگوریتم روش دسته بندی اش را انتخاب کردیم یعنی
#SVC
#این روش روش دسته بندی است
#و چون کرایه محاسباتی ستونی عددی است باید مقادیرش را دسته بندی کنیم
#تا ببینیم
#SVM
#چگونه عمل میکند

#مشخص کردن محدوده های دسته بندی
q1 = transport_df['Calculateion fare'].quantile(0.25)
q2 = transport_df['Calculateion fare'].quantile(0.50)
q3 = transport_df['Calculateion fare'].quantile(0.75)


# برش بازه‌ها
#pd.cut()
#تابعی است که یک ستون عددی رو بر اساس بازه‌هایی که خودمان مشخص می‌کنیم، دسته‌بندی می‌کند
transport_df['fare_class'] = pd.cut(
    transport_df['Calculateion fare'],
    bins=[-float('inf'), q1, q2, q3, float('inf')],
    labels=['Low', 'Medium', 'High', 'Very High']
)


transport_df.head()

```
